```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.linspace(-10, 10, 1000)
y = x**2 + 2*x + 2
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, y+1, color='red', linewidth=0.2)
ax.plot(x, y+2, color='red', linewidth=0.5)
ax.plot(x, y+2, color='red', linewidth=1.0)

ax.plot(x, y+5, color='green', linestyle=(5, (10, 3)))
ax.plot(x, y+10, color='green', linestyle='dashed')
ax.plot(x, y+15, color='green', linestyle='dashdot')

line, = ax.plot(x, y+8, color='black', lw=1.5)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, y**2, x, np.exp(x))
ax.set_title('Numbers')
ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker

ft = ticker.ScalarFormatter(useMathText=True)
ft.set_scientific(True)
ft.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(ft)
ax.yaxis.labelpad=15
ax.xaxis.lablepad=10
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))
ax[0].plot(x, x**2, x, x**3, lw=2)
ax[0].grid(color='g', alpha=0.5, linestyle='dotted', lw=0.5)

ax[1].plot(x, x**2, x, x**3, lw=2)
ax[1].grid(color='b', alpha=0.5, linestyle='dotted', lw=0.5)
```

```python
n = np.arange(1,50)
x = np.linspace(-0.5, 1.0, 50)
```

```python
fig, ax = plt.subplots(1,4, figsize=(12,3))
ax[0].scatter(x, x+.25*np.random.randn(len(x)))
ax[0].set_title('Scatter')

ax[1].step(n, n**2, lw=2)
ax[1].set_title('Step')

ax[2].bar(n, n**2, align='center', width=0.5, alpha=0.5)
ax[2].set_title('Bar')

ax[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
ax[3].set_title('Fill')
```

```python
X,Y=np.meshgrid(np.linspace(0,2*np.pi,300),np.linspace(0.2*np.pi,300))
Z=2*np.cos(X)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,2,1,projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

ax = fig.add_subplot(1,2,2, projection='3d')
p = ax.plot_surface(X,Y,Z, rstride=1, cmap=matplotlib.cm.viridis, linewidth=0, antialiased=False)
cb = fig.colorbar(p,shrink=0.5)

cset = ax.contour(X,Y,Z, zdir='z', offset=-np.pi, cmap='coolwarm')
cset = ax.contour(X,Y,Z, zdir='x', offset=-np.pi, cmap='coolwarm')
cset = ax.contour(X,Y,Z, zdir='y', offset=3*np.pi, cmap='coolwarm')
```
