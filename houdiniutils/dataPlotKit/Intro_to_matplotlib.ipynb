```python
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
import numpy as np
```

```python
X = np.arange(0, 100, 5)
Y = np.arange(100, 200, 5)
```

```python
plt.scatter(X,Y, c=Y, cmap='RdYlBu')
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
n_samples = 1000
X, y = datasets.make_circles(n_samples, noise=0.01, random_state=12)
```

```python
plt.scatter(x=X[:,0], y=X[:,1], c=y, cmap=plt.cm.RdYlBu);
```

```python
NUM_CLASSES = 4
NUM_FEATURES = 2

X_blob, y_blob = datasets.make_blobs(n_samples=2000, n_features=NUM_FEATURES, centers=NUM_CLASSES, cluster_std=1.8, random_state=32)
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.scatter(X_blob[:,0], X_blob[:,1],c=y_blob, cmap=plt.cm.RdYlBu)
```

```python
x = np.linspace(0,20,5)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```
